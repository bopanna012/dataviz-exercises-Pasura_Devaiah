```python
import pandas as pd
import plotly.express as px


# Dataset: Gapminder — GDP, Life Expectancy, Population by Country
# Source: Gapminder Foundation (gapminder.org)

df = px.data.gapminder()
print(f"Loaded: {len(df)} rows")
print(df.head())
```

```python
# explore

print(df.info())
print("Years:", sorted(df['year'].unique()))
print("Continents:", df['continent'].unique())
print(df.describe().round(1))
```

## Task 1 — Scatter: life expectancy change over time

**What to build:** A scatter showing **GDP per capita vs life expectancy** for **two years** (2000 and 2007) to show how both moved — use **colour for year** (just 2 colours), **one continent only**.

Choose any continent except Africa (that was the example). Highlight the change direction.

> 💡 Filter: `df.loc[df['continent'] == 'YOUR_CHOICE']` then filter years


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Ensure correct data types ---
df['year'] = pd.to_numeric(df['year'], errors='coerce')

# --- Filter for ONE continent (Asia) and TWO years ---
df_asia = df.loc[
    (df['continent'] == 'Asia') &
    (df['year'].isin([2000, 2007]))
].copy()

# --- Safety check (ensures both years exist) ---
print("Years in filtered data:", df_asia['year'].unique())

# --- Plot ---
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df_asia,
    x='gdpPercap',
    y='lifeExp',
    hue='year',
    palette={2000: 'lightgrey', 2007: 'blue'},  # clean + safe mapping
    s=70,
    alpha=0.8,
    legend=True
)

# --- Log scale for GDP ---
plt.xscale('log')

# --- Titles & labels (INSIGHT-BASED) ---
plt.title('Asian countries moved toward higher income and longer lives from 2000 to 2007', fontsize=13)
plt.xlabel('GDP per Capita (log scale)')
plt.ylabel('Life Expectancy')

# --- Clean styling ---
sns.despine()
plt.grid(False)

plt.show()
```

## Task 2 — Bubble chart: tell a story

**What to build:** A bubble chart (full 2007 dataset, all countries) where:
- x = GDP per capita (log scale)
- y = life expectancy
- size = population
- colour = ONE continent highlighted (your choice), all others grey
- At least one annotation explaining the highlighted group's story

> This is the grey-and-highlight technique applied to a bubble chart.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Filter safely ---
df_2007 = df[df['year'] == 2007].copy()

# --- Create highlight column ---
df_2007['color'] = df_2007['continent'].apply(
    lambda x: 'Europe' if x == 'Europe' else 'Other'
)

# --- Plot ---
plt.figure(figsize=(12,7))

sns.scatterplot(
    data=df_2007,
    x='gdpPercap',
    y='lifeExp',
    size='pop',
    hue='color',
    sizes=(20, 800),
    palette={'Europe': 'blue', 'Other': 'lightgrey'},
    alpha=0.7,
    legend=False
)

# --- Log scale ---
plt.xscale('log')

# --- Insight title ---
plt.title('Europe clusters at high income and life expectancy, highlighting global inequality (2007)', fontsize=13)

plt.xlabel('GDP per Capita (log scale)')
plt.ylabel('Life Expectancy')

# --- Annotation ---
plt.annotate(
    'European countries cluster here\n(high income, high longevity)',
    xy=(30000, 78),
    xytext=(8000, 70),
    arrowprops=dict(arrowstyle='->', color='black'),
    fontsize=10
)

# --- Clean look ---
sns.despine()
plt.grid(False)

plt.show()
```
